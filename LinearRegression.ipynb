```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
url = 'https://raw.githubusercontent.com/jarif87/DataSets/main/data.csv'
df=pd.read_csv(url)
df.head()
df.info()
df.isnull().sum()
df.describe()
plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x="x", y="y",color="purple")
plt.show()
X=df["x"]
Y=df["y"]
```

```python
m=0
c=0
l=0.0001
n=float(len(X))
iteration=1000
for i in range(iteration):
    pred=(m*X)+c
    derv_m=(-2/n)*sum(X*(Y-pred))
    derv_c=(-2/n)*sum((Y-pred))
    m = m-(l*derv_m)
    c = c-(l*derv_c)
print(m,c)
```

```python
final_pred=m*X+c

plt.figure(figsize=(6, 5))
plt.scatter(X, Y)
plt.plot(X, final_pred, color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression with Gradient Descent')

plt.show()
```
